```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/csgo_round_snapshots (1).csv')
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
  if df[col].dtypes != 'float64':
    df[col] = le.fit_transform(df[col])
```

```python
df.info()
```

```python
# Model Building

x = df.drop(columns='round_winner') # x = df.loc[:,:97]
y = df['round_winner']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=34)
```

```python
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
x_train_scaled = se.fit_transform(x_train)
x_test_scaled = se.transform(x_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(x_train_scaled,y_train)
```

```python
# x_train_lda = lda.fit_transform(x_train_scaled,y_train)
lda.transform(x_test_scaled)
```

```python
lda.coef_
```

```python
lda_coeff = np.exp(np.abs(lda.coef_))
lda_coeff
```

```python
lda_coeff = lda_coeff.flatten()
lda_coeff
```

```python
feature_names = x.columns
feature_names
```

```python
plt.figure(figsize=(15,7))
plt.bar(feature_names,lda_coeff)
plt.xticks(rotation=90)
plt.xlabel('Feature Names')
plt.ylabel('Lda coefficients')
plt.show()
```

```python
df_lda_coeff = pd.DataFrame({'feature_names' : feature_names,'feature_score' : lda_coeff})
df_lda_coeff
```

```python
# Another way of writing this is
# top_20_values = df_lda_coeff.sort_values(by='feature_score',ascending=False).head(20)
top_20_values = df_lda_coeff.nlargest(20,'feature_score')
top_20_values
```

```python
important_cols = top_20_values.index
important_cols
```

```python
x_train = x_train_scaled[:,important_cols]
x_train
```

```python
x_test = x_test_scaled[:,important_cols]
```

```python
x_test
```

```python
# Model building using Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
from sklearn.metrics import *
accuracy_score(y_test,y_pred)
```

```python
# Using Logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
# Using Decision tree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test,y_pred)
```
